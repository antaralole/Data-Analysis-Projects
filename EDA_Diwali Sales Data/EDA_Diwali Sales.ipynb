# Exploratory Data Analysis

### GENDER

#### 1) # of total males and females

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender','Amount']).size()
##ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(column)
```

```python
##group by Gender and Amount. Here I have not put as_index hence probably no error is thrown for sort function but in the below code because I am setting index values for rows as false it asks for by parameter in the sort function
df.groupby(['Gender'])['Amount'].sum().sort_values()
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

#### 2) # of total sales across males and females

```python
sales_gender = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
axi = sns.barplot(x='Gender',y='Amount',data=sales_gender)
for bars in axi.containers:
    axi.bar_label(bars)
```

### Conclusion 1: Females are greater in number compared to males and account for a major portion of the sales

### AGE

#### 1) Count of males and females across different age groups

```python
ax=sns.countplot(x='Age Group',hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ages = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

#### 2) Sales distribution  across different age groups

```python
ax=sns.barplot(x='Age Group', y ='Amount', data = sales_ages)
for bars in ax.containers:
    ax.bar_label(bars)
```

### Conclusion 2: Age groups 26-35 spend the most and are majorly females

### STATE

#### 1) Orders distribution  across states

```python
state_orders=df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
ax=sns.barplot(data=state_orders, x='State', y='Orders')
sns.set(rc={'figure.figsize':(35,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

#### 2) Sales distribution  across different states

```python
state_sales=df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sns.barplot(data=state_sales, x='State', y='Amount')
sns.set(rc={'figure.figsize':(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Conclusion 3: UP, Maharashtra and Karnataka account for majority of the orders and sales

### MARITAL STATUS

#### 1) Count of married and unmarried people

```python
sns.countplot(x='Marital_Status', data=df)
```

#### 2) Sales distribution across married/unmarried people and gender

```python
marital_sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sns.barplot(data=marital_sales, x='Marital_Status', y='Amount', hue='Gender')
sns.set(rc={'figure.figsize':(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Conclusion 4: Married females tend to buy more

### OCCUPATION

#### 1) Count of people across different occupations

```python
ax=sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

#### 2) Sales across different occupations

```python
occupation_sales=df.groupby(['Occupation','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(data=occupation_sales, x='Occupation', y='Amount')
sns.set(rc={'figure.figsize':(25,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Conclusion 5: Majority buyers are from It sector, aviation, and healthcare

### PRODUCT CATEGORY

#### 1) Distribution of products

```python
ax=sns.countplot(x='Product_Category',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

#### 2) Sales distribution across products

```python
product_sales=df.groupby(['Product_Category','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
ax=sns.barplot(data=product_sales, x='Product_Category', y='Amount')
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Conclusion 6: While clothing & apparel products are most sold they are comparitively lower in sales as against footwear & shoes which are less sold but have second highest sales
